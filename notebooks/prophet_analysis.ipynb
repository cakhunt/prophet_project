```python
import pandas as pd
import matplotlib.pyplot as plt

from prophet import Prophet
```

```python
df = pd.read_csv("../data/raw/air_passengers.csv")
df.head()
```

```python
df = df.rename(columns={
    "Month": "ds",
    "Passengers": "y"
})

df["ds"] = pd.to_datetime(df["ds"])

df.head()
```

```python
plt.figure()
plt.plot(df["ds"], df["y"])
plt.title("Air Passengers Over Time")
plt.xlabel("Date")
plt.ylabel("Passengers")
plt.show()
```

```python
model = Prophet()
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=12, freq="ME")
future.tail()
```

```python
forecast = model.predict(future)
forecast.head()
```

```python
forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail()
```

```python
fig = model.plot(forecast)
plt.show()
```

```python
fig = model.plot_components(forecast)
plt.show()
```

```python
model = Prophet()
model.add_seasonality(name="monthly", period=30.5, fourier_order=5)

model.fit(df)

future = model.make_future_dataframe(periods=12, freq="ME")
forecast = model.predict(future)

model.plot_components(forecast)
```

```python
split_index = int(len(df) * 0.8)

train_df = df.iloc[:split_index]
test_df = df.iloc[split_index:]
```

```python
model = Prophet()
model.fit(train_df)

future = model.make_future_dataframe(periods=len(test_df), freq="ME")
forecast = model.predict(future)
```

```python
pred = forecast["yhat"].tail(len(test_df)).values
actual = test_df["y"].values

plt.figure()
plt.plot(test_df["ds"], actual, label="Actual")
plt.plot(test_df["ds"], pred, label="Predicted")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(actual, pred)
print("MAE:", mae)
```
